```python
import requests
from bs4 import BeautifulSoup
import csv
from checker import *

url= 'https://www.sharesansar.com/today-share-price'
headers= {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
file= open('May_20th.csv', 'w', newline='')
writer=csv.writer(file)
writer.writerow(
    [
        'Ticker', 'Open','High','Low', 'Close', 'Volume','Change','52H'
    ]
)

r= requests.get(url,headers= headers)
soup= BeautifulSoup(r.text, 'html.parser')


table= soup.find('tbody')
# print(table)
rows= table.find_all('tr')[0:]
# print(len(rows))
for row in rows:
    ticker= row.find_all('td')[1].find('a').string
    op=row.find_all('td')[3].string
    high=row.find_all('td')[4].string
    low=row.find_all('td')[5].string
    cl=row.find_all('td')[6].string
    volume=row.find_all('td')[8].string
    chg= row.find_all('td')[12].string
    fwh= row.find_all('td')[19].string

    writer.writerow(
        [
            ticker, op, high, low, check(cl), check(volume), chg, check(fwh)
        ]
    )

file.close()
```
